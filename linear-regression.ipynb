```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('C:\\Users\\Musakalim Khan\\Downloads\\Salary.csv')
print(df.head())
df.iloc[:, -1] = df.iloc[:, -1].replace(',', '', regex=True).astype(float)
x = df.iloc[:,:-1].values # df.iloc: Access rows and columns by position (integer-based indexing).
#print(x[0:5])

y=df.iloc[:,-1].values# Syntax: df.iloc[row_index, column_index]
#print(y[0:5])

plt.scatter(x,y)
#plt.show()

from sklearn.model_selection import train_test_split
x_train, x_test , y_train , y_test = train_test_split(x,y,test_size=0.2, random_state=0)

#print(x_train.shape)
#print(x_test.shape)
#print(y_train.shape)
#print(x_test.shape)

from sklearn.linear_model import LinearRegression
model = LinearRegression()

model.fit(x_train, y_train)
#predictions = model.predict([[7.],[5.],[3.5]])

#print(predictions)
y_pred = model.predict(x_test)
print(y_pred)

#error = y_pred - y_test
#print(error)

plt.scatter(x_test, y_test)
plt.plot(x_test,y_pred, color='blue')
plt.show()

from sklearn.metrics import r2_score
r2= r2_score(y_test, y_pred)
print(r2)
```

```python
try:
    experience = float(input("Enter years of experience to predict salary: "))
    
    # Make the prediction based on user input
    user_prediction = model.predict([[experience]])
    print(f"Predicted salary for {experience} years of experience: {user_prediction[0]:.2f}")

except ValueError:
    print("Invalid input! Please enter a numeric value for experience.")
```
